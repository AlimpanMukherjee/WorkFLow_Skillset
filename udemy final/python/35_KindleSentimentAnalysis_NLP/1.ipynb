```python
import pandas as pd
data=pd.read_csv('C:/python prog/udemy final/python/35_KindleSentimentAnalysis_NLP/all_kindle_review .csv')
```

```python
data
```

```python
df=data[['rating','reviewText']]
```

```python
df['rating']=df['rating'].apply(lambda x : 0 if x<3 else 1)
```

```python
df.head()
```

```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python

clean_reviews = []

for text in df["reviewText"]:
    # 1. Remove HTML tags
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text)
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    words = word_tokenize(text)

    cleaned_words = []
    cleaned_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    clean_text = " ".join(cleaned_words)

    clean_reviews.append(clean_text)
```

```python
df['reviewText']=clean_reviews
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.20)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

bow = CountVectorizer()

X_train_bow = bow.fit_transform(X_train)   # NO .toarray()
X_test_bow  = bow.transform(X_test)

nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)

                                              
```

```python
y_pred=nb_model.predict(X_test_bow)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

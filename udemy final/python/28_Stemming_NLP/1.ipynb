## PorterStemmer

```python
words=['eating' ,'eats','eaten','eat','ate','writing','writes','programming','program','programmed']
```

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for word in words:
    print( word +'->'+stemming.stem(word))
```

#### But for some of the words the stemmingis not working properly as shown below and giving meaningless results 

```python
stemming.stem('congratulations')
```

### RegexpStemmer

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$') ### basically when it gets 'ing' or 's' or 'e' or 'able' in the last it will remove it
### and this parameter is a regular expression and needed to be passed as an argument   
reg_stemmer.stem('eating')
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
 
snowballstemmer=SnowballStemmer('english')
for word in words:
    print(word+'->'+snowballstemmer.stem(word))
```

#### Comparing the results of stemmer and snowball stemmer

```python
stemming.stem('fairly'),stemming.stem('sportingly')
```

```python
snowballstemmer.stem('fairly'),snowballstemmer.stem('sportingly')
```

```python
snowballstemmer.stem('congratulations') ### So in this case snowballstemmer will not stem the word 'congratulations'
```

#### So snowballstemmer is a little bit more accurate than porterstemmer and RegexpStemmer

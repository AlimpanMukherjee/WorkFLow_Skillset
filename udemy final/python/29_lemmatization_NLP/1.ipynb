```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
'''
POS-Noun-n
verb-v
adjective-a
adverb-r
'''
print(lemmatizer.lemmatize('dogs',pos='n'))
print(lemmatizer.lemmatize('going',pos='v')) ## now going will be treated as a verb
print(lemmatizer.lemmatize('going',pos='n')) ## now going will be treated as a noun
```

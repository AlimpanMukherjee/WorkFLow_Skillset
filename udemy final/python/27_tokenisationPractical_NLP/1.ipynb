## Tokenization
### sentence-->paragraph

```python

import nltk
nltk.download('punkt')
nltk.download("punkt_tab")  # this are downloaded because without them the new versions of nltk will not work
```

```python
from nltk.tokenize import sent_tokenize
corpus=''' hello welcome , This is Alimpan Mukherjee.
I live in Howrah and I am a Data Scientist.
I am a Python Developer.'''
document= sent_tokenize(corpus)
print(document)
```

```python
for sentence in document:
    print(sentence)
```

### paragraph-->word

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in document:
    print(word_tokenize(sentence))
    print()
```

### in this the punctuation is also treated as a seperate word

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

### In this all the full stop are attached with the last word but at the last the full stop is not attached

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```

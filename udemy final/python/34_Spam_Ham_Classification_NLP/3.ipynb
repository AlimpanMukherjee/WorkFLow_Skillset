```python
from gensim.models import Word2Vec, KeyedVectors
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
import pandas as pd
messages=pd.read_csv('spam.csv',sep='\t',names=['label','message'])
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import numpy as np
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
ps=PorterStemmer()
wn=WordNetLemmatizer()
```

## Preprocessing

Storing the preprocessed data in the corpus

```python
corpus=[]
for i in range(len(messages)):
    messages['message'][i]=messages['message'][i].lower()
    messages['message'][i]=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    messages['message'][i]=messages['message'][i].split()
    
    messages['message'][i]=[wn.lemmatize(word) for word in messages['message'][i] if word not in set(stopwords.words('english'))]
    messages['message'][i]=' '.join(messages['message'][i])

    corpus.append(messages['message'][i])
```

```python
messages
```

```python
corpus
```

## In each sentence(corpus) we do word_tokenize and store in words

```python
words=[]
for i in range(len(corpus)):
    msg=word_tokenize(corpus[i])
    words.append(msg)
```

```python
words
```

## Word2Vec


```python
import gensim
```

```python
model=gensim.models.Word2Vec(words, vector_size=200,
    window=5,
    min_count=3,
    sg=1,
    negative=10,
    epochs=20,
    workers=4)
model.wv.index_to_key
```

## We create a filtered_words to remove the empty sentences

```python
filtered_words=[]
filtered_labels=[]
for s,y in zip(words,messages.label):
    if any(word in model.wv for word in s):
        filtered_words.append(s)
        filtered_labels.append(y)
            
```

```python
words
```

```python
filtered_words
```

## We use Avg word to vec on the data based on the filtered words and store in x

```python
import numpy as np
def avg_w2v(words):
    return np.mean([model.wv[word] for word in words if word in model.wv],axis=0)
x=[]
for i in range(len(filtered_words)):
    x.append(avg_w2v(filtered_words[i]))
```

```python
x
```

```python
len(x)
```

```python
type(x)
```

### We create a array of which was previously a list

```python
x_new=np.array(x)
x_new.shape
```

## Create a Dataframe containing the avg word to vec(df)

```python
df = pd.DataFrame(x_new,columns=[f"w2v_{i}" for i in range(x_new.shape[1])])
```

```python
df
```

### Creating the output features

```python
df['output']=filtered_labels
df['output']=df['output'].map({'ham':0,'spam':1})
```

```python
df
```

```python
df.drop(df.index[0],inplace=True)
df['output']=df['output'].astype(int)
```

```python
df
```

## Training the model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('output',axis=1),df['output'],test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model=model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score
print(accuracy_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
```

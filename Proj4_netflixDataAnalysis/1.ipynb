### how many movies are there in each country every year

```python
country_df=df[['country','year_added']].dropna().copy()
country_df['country']=country_df['country'].str.split(',')

country_df = country_df.explode('country')
country_df['country'] = country_df['country'].str.strip()
country_df
```

```python
data_over_years=country_df.groupby(['year_added','country']).size().reset_index(name='count')
data_over_years
```

```python
country_df[(country_df['country']=='India') &  (country_df['year_added']==2021)].size
```

```python
def plot_by_year(country):
    temp=data_over_years[data_over_years['country']==country]
    plt.plot(temp['year_added'],temp['count'],marker='o',label=country)
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.title(f'Count of {country} by Year')
    plt.legend()
    plt.show()
    
plot_by_year('United States')
plot_by_year('India')
plot_by_year('Denmark')
```

### Growth Rate of the countries over the years

```python
data_over_years['growth']=data_over_years.groupby('country')['count'].pct_change()
data_over_years[data_over_years['country']=='India']
```

```python
data_over_years[data_over_years['country']=='Denmark']
```

```python
plt.figure(figsize=(11, 8))
def percent_change_by_year(country,color):
    temp=data_over_years[data_over_years['country']==country]
    
    plt.plot(temp['year_added'],temp['growth'],marker='o',label=country,color=color)
    
    plt.xlabel('Year')
    plt.ylabel('Growth')
    plt.legend()
    ##plt.xticks(sorted(set('Years')))
    # plt.show()

percent_change_by_year('United States','red')
percent_change_by_year('India','orange')
percent_change_by_year('Denmark','black')
percent_change_by_year('France','blue')
percent_change_by_year('United Kingdom','pink')
```
